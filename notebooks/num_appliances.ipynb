```python
from __future__ import print_function
from nilmtk.dataset import DataSet
from nilmtk.cross_validation import train_test_split
from nilmtk.disaggregate.fhmm_exact import FHMM
from nilmtk.disaggregate.co_1d import CO_1d
from nilmtk.metrics import rms_error_power
from nilmtk.metrics import mean_normalized_error_power
from nilmtk.sensors.electricity import Measurement
from nilmtk.stats.electricity.building import top_k_appliances
from nilmtk.preprocessing.electricity.building import filter_top_k_appliances
from nilmtk.preprocessing.electricity.building import downsample
from nilmtk.preprocessing.electricity.building import filter_datetime
from nilmtk.dataset import DataSet
import time
import pandas as pd
import matplotlib.pyplot as plt
import resource
import os
```

```python
dataset_names=["ampds"]
```

```python
base_path="/home/nipun/Dropbox/nilmtk_datasets/"
```

```python
DISAGG_FEATURE = Measurement('power', 'active')
```

```python
dataset =DataSet()
dataset.load_hdf5(os.path.join(base_path, dataset_names[0]))
```

```python
disaggregators={}
disaggregators['fhmm']=FHMM()
disaggregators['co']=CO_1d()
```

```python
train_time={'co':{},'fhmm':{}}
disaggregate_time={'co':{},'fhmm':{}}
re={'co':{},'fhmm':{}}
mne={'co':{},'fhmm':{}}
```

```python
for num_appliances in range(3,7):
    print("\n\nNumber of appliances is: {}".format(num_appliances))
    b = dataset.buildings[1]
    
    # Filter only a month to speed up computations
    b = filter_datetime(b, '08-01-2012', '09-01-2012')
      
    # Filtering to include only top `num_appliances` appliances
    b = filter_top_k_appliances(b, num_appliances)

    # Dividing the data into train and test
    train, test = train_test_split(b) 
    
    
    for disaggregator_name, disaggregator in disaggregators.iteritems():   
        # Train
        t1 = time.time()
        disaggregator.train(train, disagg_features=[DISAGG_FEATURE])
        t2 = time.time()
        print("Runtime to train for {} = {:.2f} seconds".format(disaggregator_name, t2 - t1))
        train_time[disaggregator_name][num_appliances]=t2-t1
    
        # Disaggregate
        t1 = time.time()
        disaggregator.disaggregate(test)
        t2 = time.time()
        print("Runtime to disaggregate for {}= {:.2f} seconds".format(disaggregator_name, t2 - t1))
        disaggregate_time[disaggregator_name][num_appliances]=t2-t1
    
        # Metrics
        predicted_power = disaggregator.predictions
        app_ground = test.utility.electric.appliances
        ground_truth_power = pd.DataFrame({appliance: app_ground[appliance][DISAGG_FEATURE] for appliance in app_ground})

        # RMS Error
        re[disaggregator_name][num_appliances] = rms_error_power(predicted_power, ground_truth_power)

        # Mean Normalized Error
        mne[disaggregator_name][num_appliances] = mean_normalized_error_power(predicted_power, ground_truth_power)    
    
```

```python
def dict_to_df(dictionary):
    series={}
    for algo_name, val in dictionary.iteritems():
        series[algo_name] = pd.Series(val , val.keys())
    return pd.DataFrame(series)
```

#### Comparing training time

```python
train_time_df = dict_to_df(train_time)
ax = train_time_df.plot(kind='bar')
ax.set_ylabel("Time (in s)");
ax.set_xlabel("# Appliances");
```

#### Comparing disaggregration time

```python
disaggregate_time_df = dict_to_df(disaggregate_time)
ax=disaggregate_time_df.plot(kind='bar')
ax.set_ylabel("Time (in s)");
ax.set_xlabel("# Appliances");
```

### Comparing disaggregation results within algorithms when N is varied

```python
for num_appliances in range(3,7):
    for algo_name, results in mne.iteritems():
        d = dict_to_df(results)
        plt.figure()
        d.plot(kind='bar',title="{} for N = {}".format(algo_name, num_appliances))
```
