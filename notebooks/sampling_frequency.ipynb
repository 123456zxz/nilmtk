```python
from __future__ import print_function
from nilmtk.dataset import DataSet
from nilmtk.cross_validation import train_test_split
from nilmtk.disaggregate.fhmm_exact import FHMM
from nilmtk.disaggregate.co_1d import CO_1d
from nilmtk.metrics import rms_error_power
from nilmtk.metrics import mean_normalized_error_power
from nilmtk.sensors.electricity import Measurement
from nilmtk.stats.electricity.building import top_k_appliances
from nilmtk.preprocessing.electricity.building import filter_top_k_appliances
from nilmtk.preprocessing.electricity.building import downsample
from nilmtk.dataset import DataSet
import time
import pandas as pd
import matplotlib.pyplot as plt
import resource
import os
```

```python
# Setting the limits to 5 GB RAM usage
megs = 5000
resource.setrlimit(resource.RLIMIT_AS, (megs * 1048576L, -1L))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset_names=["ampds"]
```

```python
base_path="/home/nipun/Dropbox/nilmtk_datasets/"
```

```python
DISAGG_FEATURE = Measurement('power', 'active')
```

```python
dataset =DataSet()
dataset.load_hdf5(os.path.join(base_path, dataset_names[0]))
```

```python
# Experiment on first (and only) building
b = dataset.buildings[1]
```

```python
train_time_fhmm={}
disaggregate_time_fhmm={}
re_fhmm={}
mne_fhmm={}
train_time_co={}
disaggregate_time_co={}
re_co={}
mne_co={}
```

```python
for downsampling_freq in ['1T','2T','5T','10T','15T']:
    print(downsampling_freq)
    b = dataset.buildings[1]
    
    # Downsample
    b = downsample(b, downsampling_freq)
    
    # Filtering to include only top 5 appliances
    b = filter_top_k_appliances(b, 5)

    # Dividing the data into train and test
    train, test = train_test_split(b)
        
    # Initializing FHMM Disaggregator
    disaggregator = FHMM()
    
    # Train
    t1 = time.time()
    disaggregator.train(train, disagg_features=[DISAGG_FEATURE])
    t2 = time.time()
    print("Runtime to train = {:.2f} seconds".format(t2 - t1))
    train_time_fhmm[downsampling_freq]=t2-t1
    
    # Disaggregate
    t1 = time.time()
    disaggregator.disaggregate(test)
    t2 = time.time()
    print("Runtime to disaggregate = {:.2f} seconds".format(t2 - t1))
    disaggregate_time_fhmm[downsampling_freq]=t2-t1
    
    # Metrics
    predicted_power = disaggregator.predictions
    app_ground = test.utility.electric.appliances
    ground_truth_power = pd.DataFrame({appliance: app_ground[appliance][DISAGG_FEATURE] for appliance in app_ground})

    # RMS Error
    re_fhmm[downsampling_freq] = rms_error_power(predicted_power, ground_truth_power)

    # Mean Normalized Error
    mne_fhmm[downsampling_freq] = mean_normalized_error_power(predicted_power, ground_truth_power)
    
    # Initializing FHMM Disaggregator
    disaggregator = CO_1d()
    
    # Train
    t1 = time.time()
    disaggregator.train(train, disagg_features=[DISAGG_FEATURE])
    t2 = time.time()
    print("Runtime to train = {:.2f} seconds".format(t2 - t1))
    train_time_co[downsampling_freq]=t2-t1
    
    # Disaggregate
    t1 = time.time()
    disaggregator.disaggregate(test)
    t2 = time.time()
    print("Runtime to disaggregate = {:.2f} seconds".format(t2 - t1))
    disaggregate_time_co[downsampling_freq]=t2-t1
    
    # Metrics
    predicted_power = disaggregator.predictions
    app_ground = test.utility.electric.appliances
    ground_truth_power = pd.DataFrame({appliance: app_ground[appliance][DISAGG_FEATURE] for appliance in app_ground})

    # RMS Error
    re_co[downsampling_freq] = rms_error_power(predicted_power, ground_truth_power)

    # Mean Normalized Error
    mne_co[downsampling_freq] = mean_normalized_error_power(predicted_power, ground_truth_power)
    
```
