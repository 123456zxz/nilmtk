```python
from __future__ import print_function
import os
import matplotlib.pyplot as plt
import time
```

```python
from nilmtk.dataset import DataSet
```

```python
dataset_names=["redd/low_freq","ampds","pecan_1min","pecan_15min","iawe"]
```

```python
base_path="/home/nipun/Dropbox/nilmtk_datasets/"
```

```python
for dataset_name in dataset_names:
    dataset_path = os.path.join(base_path, dataset_name)
    dataset = DataSet()
    dataset.load_hdf5(dataset_path)
    print("Loaded dataset")
    for building_number, building in dataset.buildings.iteritems():
        fig = plt.figure(figsize=(12,8))  
        start_time = time.time()
        building.utility.electric.plot_missing_samples()
        end_time = time.time()
        print("Time taken for plotting missing samples in building {} of dataset {} ={:2f} seconds".format(building_number, dataset_name, end_time-start_time))
    print("*"*80)
```
