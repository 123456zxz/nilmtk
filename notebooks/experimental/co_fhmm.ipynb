```python
disag_filename = join(data_dir, 'redd-disag-fhmm-new.h5')
output = HDFDataStore(disag_filename, 'w')
fhmm.disaggregate(test_elec.mains(), output, sample_period=60)
output.close()
```

```python
disag_filename = join(data_dir, 'redd-disag-fhmm-new.h5')
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec
```

```python
disag_elec.plot()
```

```python
f1 = f1_score(disag_elec, test_elec)
f1.index = disag_elec.get_labels(f1.index)
f1.plot(kind='barh')
plt.ylabel('appliance');
plt.xlabel('f-score');
plt.title("FHMM");
```

```python
# Train
co = CombinatorialOptimisation()
co.train(train_elec)
```

```python
# Disaggregate
disag_filename = join(data_dir, 'redd-disag-co.h5')
output = HDFDataStore(disag_filename, 'w')
co.disaggregate(test_elec.mains(), output)
output.close()
```

## Examine disaggregated data

#### CO

```python
disag_filename = join(data_dir, 'wikienergy-disag-co.h5')
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec
disag_elec.plot()
disag.store.close()
plt.title("CO");
```

## Calculate accuracy of disaggregation

```python
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec

f1 = f1_score(disag_elec, elec)
f1.index = disag_elec.get_labels(f1.index)
f1.plot(kind='barh')
plt.ylabel('appliance');
plt.xlabel('f-score');
plt.title("CO");

disag.store.close()
```

```python
disag_filename = join(data_dir, 'wikienergy-disag-fhmm.h5')
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec
disag_elec.plot()
plt.title("FHMM");
disag.store.close()
```

```python
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec

f1 = f1_score(disag_elec, elec)
f1.index = disag_elec.get_labels(f1.index)
f1.plot(kind='barh')
plt.ylabel('appliance');
plt.xlabel('f-score');
plt.title("FHMM");

disag.store.close()
```
