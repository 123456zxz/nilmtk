```python
from IPython.display import display
from IPython.core.display import HTML

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import nilmtk
from nilmtk import DataSet, TimeFrame
from nilmtk.disaggregate import CombinatorialOptimisation
from nilmtk import HDFDataStore
from nilmtk.metrics import f1_score

import warnings
warnings.filterwarnings("ignore")

def pretty_print_dict(dictionary):
    html = '<ul>'
    for key, value in dictionary.iteritems():
        html += '<li><strong>{}</strong>: '.format(key)
        if isinstance(value, list):
            html += '<ul>'
            for item in value:
                html += '<li>{}</li>'.format(item)
            html += '</ul></li>'
        else:
            html += '{}</li>'.format(value)
    html += '</ul>'
    display(HTML(html))
```

## Loading data

```python
data_dir = '/Users/user/data/'
we = DataSet(data_dir + 'wikienergy.h5')
print('loaded ' + str(len(we.buildings)) + ' buildings')
```

## Examine dataset metadata

```python
pretty_print_dict(we.metadata)
```

## Examine metadata for a single house

```python
building_number = 11
pretty_print_dict(we.buildings[building_number].metadata)
```

## Examine sub-metered appliances

```python
elec = we.buildings[building_number].elec
elec.appliances
```

## Plot sub-metered data for a single day

```python
plt.figure(figsize=(16,8))
plt.clf()
we.store.window = TimeFrame(start='2014-04-01 00:00:00-05:00', end='2014-04-02 00:00:00-05:00')
elec.plot()
```

## Training and disaggregation

```python
co = CombinatorialOptimisation()
'''
model = {
 2: np.array([   0,  350]),
 3: np.array([ 0, 100]),
 4: np.array([ 0, 3000]),
}
co.model = model
'''
co.train(elec)
baseline = elec.mains().power_series_all_data().min()
co.model[12] = np.array([baseline]) # add baseline

disag_filename = data_dir + 'wikienergy-disag.h5'
output = HDFDataStore(disag_filename, 'w')
co.disaggregate(elec.mains(), output)
output.close()
```

## Examine learned model

```python
pretty_print_dict(co.model)
```

## Examine disaggregated data

```python
disag = DataSet(disag_filename)
plt.figure(figsize=(16,8))
disag_elec = disag.buildings[building_number].elec
disag_elec.plot()
disag.store.close()
```

## Calculate accuracy of disaggregation

```python
disag = DataSet(disag_filename)
disag_elec = disag.buildings[building_number].elec
disag_elec.meters = disag_elec.meters[0:-1] # remove baseline

f1 = f1_score(disag_elec, elec)
print(f1)

disag.store.close()
```
