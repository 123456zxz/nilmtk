Considering top 5 appliances for disaggregation

```python
#building = prep_b.filter_top_k_appliances(building, k=6)
building =prep_b.filter_contribution_less_than_x(building)
```

```python
from nilmtk.disaggregate.fhmm_exact import FHMM
```

```python
disaggregator = FHMM()
```

```python
from nilmtk.cross_validation import train_test_split
```

```python
train, test =train_test_split(building, train_split=0.5)
```

```python
DISAGG_FEATURE = Measurement('power','active')
```

```python
disaggregator.train(train)
```

```python
disaggregator.disaggregate(test)
```

```python
predicted_power = disaggregator.predictions
app_ground = test.utility.electric.appliances
ground_truth_power = pd.DataFrame({appliance: app_ground[appliance][DISAGG_FEATURE] for appliance in app_ground})
```

```python
predicted_power[('dehumidifier',1)]
```

```python
fig, axes = plt.subplots(ncols=2, sharex=True)
predicted_power[('dehumidifier',1)].plot(title='Predicted power\nFHMM',ax=axes[0])
ground_truth_power[('dehumidifier',1)].plot(ax=axes[1])
```

```python
from nilmtk.metrics import mean_normalized_error_power
```

```python
mean_normalized_error_power(predicted_power, ground_truth_power)
```
